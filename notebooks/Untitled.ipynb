```python
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(-50.,50.).reshape((10,10))
np.fill_diagonal(X, np.nan)
print(X)
```

```python
plt.pcolormesh(X, cmap='PuOr_r', vmin=-50, vmax=50)
plt.colorbar()
```

```python
x = np.linspace(0, 1, 256)
plt.plot(x, plt.get_cmap('PuOr_r')(x))
nan_val = plt.get_cmap('PuOr_r')(np.nan)
for v in nan_val:
   plt.plot(0.5, v, 'o', mfc=nan_val, mec='k')
```

```python
plt.pcolormesh(np.arange(256)[:,None], cmap='PuOr_r', alpha=None)
```

```python
fig = plt.figure(figsize=(2,4))
ax = fig.add_axes([0,0,0.2,1])
ax.imshow(np.arange(256)[::-1,None], cmap='PuOr_r', aspect=20 * 1/256)
ax.set_xticks([])
ax.set_yticks([])

ax = fig.add_axes([0.5,0,0.1,1])
ax.pcolormesh(np.arange(256)[:,None], cmap='PuOr_r')
ax.set_xticks([])
ax.set_yticks([])
#plt.colorbar(ticks=[])
```
